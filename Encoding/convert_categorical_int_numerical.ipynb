#1. one hot technique using sklearn


```python
#import library for preprocessing
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
```

```python
ohetranform=ohe.fit_transform(df[['fuel','owner']])
```

```python
print(ohetranform)
```

```python

df.head()
```

```python
#to show data in correct order
ohetranform.head()
```

#2. K-1 one hot encoding

```python
#add this data to our df dataset
df=pd.concat([df,ohetranform],axis=1).drop(columns=['fuel','owner'])
```
